```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.pandas.set_option('display.max_columns',None)
```

```python
data=pd.read_csv('Downloads/train.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
features_with_nan=[i for i in data.columns if data[i].isnull().sum()>1]
```

```python
for i in features_with_nan:
    print(i,np.round(data[i].isnull().mean(),4),'% missing value')
```

# finding relation b/w missing value in salesprice

```python
for i in features_with_nan:
    df=data.copy()
    df[i]=np.where(df[i].isnull(),1,0)
    df.groupby(i)['SalePrice'].median().plot.bar()
    plt.title(i)
    plt.show()
    
```

### Numerical VAriables

```python
Num_var=[i for i in data.columns if data[i].dtypes !='O']
print('No. of Num VAr {}'.format(len(Num_var)))
data[Num_var].head()
```

## Temporal VARiables(eg datetime var)

```python
Year_feature=[i for i in Num_var if 'Yr' in i or 'Year' in i]
Year_feature
```

```python
for i in Year_feature:
    print(i,data[i].unique())
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year')
plt.ylabel('Median PRice')
plt.title('Price vs Year sold')
```

```python
for i in Year_feature:
    if i !='YrSold':
        df=data.copy()
        df[i]=df['YrSold']-df[i]
        plt.scatter(df[i],df['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('SalesPrice')
        plt.show()
```

```python
discrete_features=[i for i in Num_var if len(df[i].unique())<25 and i not in Year_feature+['Id']]
print('No. of discrete Featrues are {}'.format(len(discrete_features)))
```

```python
discrete_features
```

```python
for i in discrete_features:
    df=data.copy()
    df.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('salesPrice')
    plt.show()
```

```python
continuos_var=[i for i in Num_var if i not in discrete_features+Year_feature+['Id']]
print('Total No of Continous Features are {}'.format(len(continuos_var)))
```

```python
continuos_var
```

```python
for i in continuos_var:
    df=data.copy()
    data[i].hist(bins=25)
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

# Convert the non gaussian into gaussian or normal distribution

## we will be using LOg Transfromation

```python
for i in continuos_var:
    df=data.copy()
    if 0 in df[i].unique():
        pass
    else:
        df[i]=np.log(df[i])
        df['salePrice']=np.log(df['SalePrice'])
        plt.scatter(df[i],df['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('Sales PRice')
        plt.show()
    
```

# OUTLIERS

```python
for i in continuos_var:
    df=data.copy()
    if 0 in df[i].unique():
        pass
    else:
        df[i]=np.log(df[i])
        df.boxplot(column=i)
        plt.ylabel(i)
        plt.title(i)
        plt.show()
    
```
